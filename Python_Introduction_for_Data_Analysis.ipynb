```python
import pandas as pd
```

```python
df = pd.read_csv('superstore.csv')
```

```python
print(df)
```

1. Produk apa yang paling laku?

```python
highest_sale_product = ''
highest_sale_amount = 0

for i in range(len(df)):
  amount = df.loc[i, 'Quantity']
  product = df.loc[i, 'Product Name']

  if amount > highest_sale_amount:
    highest_sale_product = product
    highest_sale_amount = amount

print("Produk paling laku:", highest_sale_product)
print("Quantity:", highest_sale_amount)
```

2. Produk apa yang paling menguntungkan dan merugikan?

```python
profit_product = ''
deficit_product = ''

profit_amount = 0
deficit_amount = 0

for i in range(len(df)):
  amount = df.loc[i, 'Profit']
  product = df.loc[i, 'Product Name']

  if amount > profit_amount:
    profit_product = product
    profit_amount = amount
  elif amount < deficit_amount:
    deficit_product = product
    deficit_amount = amount

print("Produk paling menguntungkan:", profit_product)
print("Profit:", profit_amount)

print("Produk paling merugikan:", deficit_product)
print("Deficit:", deficit_amount)
```

```python
def min_and_max(df, column_name1, column_name2):
  profit_product = ''
  deficit_product = ''
  profit_amount = 0
  deficit_amount = 0

  for i in range(len(df)):
    amount = df.loc[i, column_name1]
    product = df.loc[i, column_name2]

    if amount > profit_amount:
      profit_product = product
      profit_amount = amount
    elif amount < deficit_amount:
      deficit_product = product
      deficit_amount = amount

  return profit_product, profit_amount, deficit_product, deficit_amount

min_and_max(df, 'Profit', 'Product Name')

print("Produk paling menguntungkan:", profit_product)
print("Profit:", profit_amount)

print("Produk paling merugikan:", deficit_product)
print("Deficit:", deficit_amount)
```

3. Berapa banyak produk yang mengalami kerugian?

```python
loss_count = 0

for i in range(len(df)):
  amount = df.loc[i, 'Profit']

  if amount < 0:
    loss_count += 1

print("Banyaknya produk yang mengalami kerugian:", loss_count)
```
